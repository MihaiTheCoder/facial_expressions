```python
import pandas as pd
import random
```

```python
from PIL import Image
from IPython.display import display, Image
```

```python
df_legend = pd.read_csv('data\legend.csv')
```

```python
not_neutral = df_legend[~df_legend.emotion.str.lower().isin(['neutral', 'happiness'])]
```

```python
def get_random_image(df):
    random_x = df.sample(n=1)
    image_path = 'images/' + list(random_x.image)[0]
    return (list(random_x.emotion)[0], image_path)
    
```

```python
rand_img = get_random_image(not_neutral)
display(Image(filename=rand_img[1]))
```
